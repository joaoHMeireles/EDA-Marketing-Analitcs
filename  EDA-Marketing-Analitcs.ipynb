##### 2.1.1 Informações de educação

```python
education_columns_info = {}
for column in df.columns:
    if(column.find('education') != -1):
        education_columns_info[column[10:]] = df[column]

education_properties = pd.DataFrame(education_columns_info)
education_properties = education_properties[["Basic", "2n Cycle","Graduation", "Master","PhD"]]
education_properties
```

```python
education_levels = education_properties.columns
quantity_education_levels = education_properties.apply(lambda x: x.value_counts().loc[1])

quantity_education_levels
info_plot = pd.DataFrame({'Níveis de escolarização': education_levels, 'Quantidade de pessoas': quantity_education_levels})

plt.gcf().set_size_inches(10,5)
sns.barplot(data=info_plot, x="Níveis de escolarização", y="Quantidade de pessoas", palette=['b']).set(title='Número de pessoas para cada nível de escolarização')
```

##### 2.1.2 Informações de matrimônio

```python
marital_columns_info = {}
for column in df.columns:
    if(column.find('marital') != -1):
        marital_columns_info[column[8:]] = df[column]

marital_properties = pd.DataFrame(marital_columns_info)
marital_properties = marital_properties[["Single", "Together", "Married", "Divorced", "Widow"]]
marital_properties
```

```python
marital_status = marital_properties.columns
quantity_marital_status = marital_properties.apply(lambda x: x.value_counts().loc[1])

# fig = plt.figure(figsize =(8,5))
plt.bar(marital_status, quantity_marital_status)
plt.show()
```
