##### 2.2 Informações de educação

```python
education_columns_info = {}
for column in df.columns:
    if(column.find('education') != -1):
        education_columns_info[column] = df[column]

education_properties = pd.DataFrame(education_columns_info)
education_properties
```

```python
education_levels = education_properties.columns
quantity_education_levels = education_properties.value_counts()

fig = plt.figure(figsize =(10, 7))
plt.bar(education_levels, quantity_education_levels)
plt.show()
```

##### 2.3 Informações de matrimônio

```python
marital_columns_info = {}
for column in df.columns:
    if(column.find('marital') != -1):
        marital_columns_info[column] = df[column]

marital_properties = pd.DataFrame(marital_columns_info)
marital_properties
```

```python
marital_status = marital_properties.columns
quantity_marital_status = marital_properties.value_counts()

fig = plt.figure(figsize =(10, 7))
plt.bar(marital_status, quantity_marital_status)
plt.show()
```
