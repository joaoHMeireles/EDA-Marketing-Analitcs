```python
education_columns_info = {}
for column in df.columns:
    if(column.find('education') != -1):
        education_columns_info[column] = df[column]

education_properties = pd.DataFrame(education_columns_info)
education_properties
```

```python
marital_columns_info = {}
for column in df.columns:
    if(column.find('marital') != -1):
        marital_columns_info[column] = df[column]

marital_properties = pd.DataFrame(marital_columns_info)
marital_properties
```

```python
amount_columns_info = {}
for column in df.columns:
    if(column.find('Mnt') != -1):
        amount_columns_info[column] = df[column]

amount_properties = pd.DataFrame(amount_columns_info)
amount_properties
```

```python
accecpted_campaigns_columns_info = {}
for column in df.columns:
    if(column.find('AcceptedCmp') != -1):
        accecpted_campaigns_columns_info[column] = df[column]

accecpted_campaigns_properties = pd.DataFrame(accecpted_campaigns_columns_info)
accecpted_campaigns_properties
```
